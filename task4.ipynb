**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    kernel = numpy.array ( [[-1,0,1],
                            [-1,0,1],
                            [-1,0,1]])

    result = numpy.zeros(img_in.shape)
    size = 3
    n  = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape): #loop over all pixels
        if n<= p[0] < img_in.shape[0]-n and n<= p[1] < img_in.shape[1]-n:
            neighborhood_sum = 0
            for q in numpy.ndindex(size, size): #nachbarschaft
               
                
                i= p[0]+q[0]-n
        
                j=p[1]+q[1] -n
                
                weight = kernel [q[0], q[1]]
                neighborhood_sum = neighborhood_sum + img_in[i,j]* weight
                result[p]= neighborhood_sum/6.0
            
    return result

    
def prewitt_v(img_in):
    kernel = numpy.array ([[-1,-1,-1],
                            [0,0,0],
                            [1,1,1]])

    result = numpy.zeros(img_in.shape)
    size = 3
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape): #loop over all pixels
        if n<= p[0] < img_in.shape[0]-n and n<= p[1] < img_in.shape[1]-n:
            neighborhood_sum = 0
            for q in numpy.ndindex(size, size): #nachbarschaft
               
                
                i= p[0]+q[0]-n
                
                j=p[1]+q[1] -n
                
                weight = kernel [q[0], q[1]]
             
                neighborhood_sum = neighborhood_sum + img_in[i,j]* weight
                result[p]= neighborhood_sum/6.0
            

    return result
```
